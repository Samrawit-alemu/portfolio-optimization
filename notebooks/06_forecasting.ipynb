```python
import sys
sys.path.append("../")

from src.forecasting import *
```

```python
TICKERS = [
    "TSLA",
    "SPY",
    "BND"
]
```

```python
for ticker in TICKERS:

    print("="*80)

    print(ticker)

    df = load_dataset(ticker)

    train, test = split_data(df)

    model = train_arima(train["Close"])

    prediction, conf = predict_test(
        model,
        test["Close"]
    )

    evaluate(
        test["Close"],
        prediction
    )

    plot_test_prediction(
        train["Close"],
        test["Close"],
        prediction,
        ticker
    )

    future, future_conf = forecast_future(
        model,
        periods=252
    )

    plot_future_forecast(
        df,
        future,
        future_conf,
        ticker
    )

    expected_return = summarize_forecast(
        future
    )

    print(
        analyze_trend(future)
    )

    print(
        confidence_analysis(
            future_conf
        )
    )

    print(
        market_insight(
            expected_return
        )
    )

    print(
        market_risk(
            future_conf
        )
    )
```
